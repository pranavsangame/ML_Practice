# One hot encoder

```python
one_hot_enco = OneHotEncoder()
array = one_hot_enco.fit_transform(df2[['num-of-doors']]).toarray()
num_of_door = pd.DataFrame(array,columns=sorted(df2['num-of-doors'].unique()))
num_of_door
```

# getdummies

```python
df2.get('num-of-cylinders')
```

```python
df_get_dum = pd.get_dummies(df2,columns=['num-of-cylinders','num-of-doors'],drop_first=True)
df_get_dum
```

```python
df1.info()
```

```python
x = df1.drop('price',axis='columns')
y = df1.price
```

```python
df1.dtypes
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.15,random_state=3)
```

```python
linear_model = LinearRegression()
linear_model
```

```python
linear_model.fit(x_train,y_train)
```

# residual

```python
y_pred = linear_model.predict(x_test)
print('mse is - ',mean_squared_error(y_test,y_pred))
print('r2 score is - ',r2_score(y_test,y_pred))
```

```python
x = df_get_dum.drop('price',axis='columns')
y = df_get_dum.price
```
