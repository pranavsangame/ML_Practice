```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('titanic.csv')
df1 = df.copy()
df2 = df.copy()
df
```

### min-max scalar

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
normal_scalar= MinMaxScaler()
scaled_fair = normal_scalar.fit_transform(df[['Fare']])
#print(scaled_fair)
df['Fare'] = pd.DataFrame(scaled_fair,columns={'Fare'})
df
```

```python
df.info()
```

```python
df['Fare'].unique()
```

## Standard scalar

```python
df1
```

```python
from sklearn.preprocessing import StandardScaler
std_scalar = StandardScaler()
scaled_values = std_scalar.fit_transform(df1[['Fare']])
df1['Fare'] = pd.DataFrame(scaled_values)
df1
```

```python
df1['Fare'].describe()
```

```python
df1.Fare.min(),df1.Fare.max()
```

```python
import seaborn as sns
sns.boxplot(df['Fare'])
```

## Robust scalar

```python
from sklearn.preprocessing import RobustScaler

Rob_scaler = RobustScaler()
rob_scaled_values = Rob_scaler.fit_transform(df2[['Fare']])
df2['Fare']=pd.DataFrame(rob_scaled_values)
df2
```

```python
df2['Fare'].describe()
```

# Skew correction

## log transform

```python
df = pd.read_csv('titanic.csv')
df
```

```python
df.dropna(subset=['Age'],inplace=True)
```

```python
sns.kdeplot(df['Age'])
```

```python
df['Age']=np.log(df['Age'])
df['Age']
```

```python
sns.kdeplot(df['Age'])
```

## square root transform

```python
df = pd.read_csv('titanic.csv')
df['Age']= np.sqrt(df['Age'])
sns.kdeplot(df['Age'],fill=True)
```

## cube root trandform

```python
df = pd.read_csv('titanic.csv')
df['Age']= np.cbrt(df['Age'])
sns.kdeplot(df['Age'],fill=True)
```
