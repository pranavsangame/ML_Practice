## Calculation of Information Gain:


```python
how is the root node decided? how does the best split happen at the root node? 

for the same, we calculate the Information gain of the entire data.
The feature having high information gain will be decided as the splitting factor at the root node.
```

```python
I.G = E(S) - sum[Weighted Avg* Entropy of each unique aspect of the feature]

where,
I.G >> Information gain
E(S)>> Overall Entropy

Entropy >> It is used to find the major of impurity of node.
I.G     >> It is measurment of change  in entropy
```

```python
#formula for entropy
Entropy= -(Prob Y)*np.log2(Prob Y) - (Prob N)*np.log2(Prob N)
```
